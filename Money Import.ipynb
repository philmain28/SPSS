```python
places = {}
count = 0
for place in Description:    
    if place in places:
        #places['thing'] = count
        places[place].append(count)
    else:
        places[place] = [count]
        #places['thing'] = count
    count +=1 
```

```python
moneys = [Amount[ind] for ind in places['\xa0CARD PAYMENT TO FRESH']]
keys = [key for key in places.keys()]
bar_data = {}
for alicia in keys:
    bar_data[alicia] = np.sum(np.array([Amount[ind] for ind in places[alicia]]))

bars = sorted(bar_data.values())
labels = sorted(bar_data, key = bar_data.get)

money_in = []
money_in_label = []
money_out = []
money_out_label = []
for bar, label in list(zip(bars,labels)):
    if bar < 0:
        money_out.append(abs(bar))
        money_out_label.append(label) 
    else:     
        money_in.append(abs(bar))
        money_in_label.append(label)

money_spent = dict(zip(money_out_label, money_out))
money_earned = dict(zip(money_in_label, money_in))        
        
        
# Now lets make some cool pie charts

pies = { 
         'card_label'      : [],
         'card_value'      : [], 
         'cash_label'      : [], 
         'cash_value'      : [],
         'transfer_label'  : [], 
         'transfer_value'  : []   }

for spend in money_spent:
    if "CARD" in spend:
        the_place = spend[len('\xa0CARD PAYMENT TO '):]
        pies['card_label'].append(the_place)
        pies['card_value'].append(money_spent[spend])
    elif "CASH" in spend:
        the_place = spend[len('\xa0CASH WITHDRAWAL AT '):]
        pies['cash_label'].append(the_place)
        pies['cash_value'].append(money_spent[spend])
    elif "STANDING" in spend:
        the_place = spend[len('\xa0STANDING ORDER VIA FASTER PAYMENT TO '):]
        pies['transfer_label'].append(the_place)
        pies['transfer_value'].append(money_spent[spend])
    elif "DIRECT DEBIT" in spend:
        the_place = spend[len('\xa0DIRECT DEBIT PAYMENT TO'):]
        pies['transfer_label'].append(the_place)
        pies['transfer_value'].append(money_spent[spend])


card_args = np.argsort(pies['card_value'])
temp_card = []
temp_card_label = []
for i in range(1, np.size(card_args)+1):
    temp_card.append(pies['card_value'][card_args[-i]])
    temp_card_label.append(pies['card_label'][card_args[-i]])

pies['card_value'] = temp_card 
pies['card_label'] = temp_card_label

card_args = np.argsort(pies['transfer_value'])
temp_card = []
temp_card_label = []
for i in range(1,np.size(card_args)+1):
    temp_card.append(pies['transfer_value'][card_args[-i]])
    temp_card_label.append(pies['transfer_label'][card_args[-i]])

pies['transfer_value'] = temp_card 
pies['transfer_label'] = temp_card_label


```
